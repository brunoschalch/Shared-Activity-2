```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
import random
%matplotlib inline
from numpy import genfromtxt
from keras import Sequential
from keras import callbacks
from keras.layers import Dense
from keras.initializers import glorot_normal
from keras.optimizers import SGD
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import normalize
```

```python
AdMob = pd.read_csv("AdMob Real World Data Android 20173.csv")
AdMob
```

```python
# For tensorboard
tbCallBack = callbacks.TensorBoard(log_dir='./Graph', histogram_freq=0, write_graph=True, write_images=True)

results = np.array(AdMob)

#results.remove(results[pd.isnull(results)])

X = results[:, [1, 2, 3, 6, 7, 10]]
Y = results[:, [16]]

# This removes al NaNs (Keras may handle this)
X[pd.isnull(X)] = 0
Y[pd.isnull(Y)] = 0

#Normalize
#X = X / np.linalg.norm(X)
Y = Y / np.linalg.norm(Y)
X= X / X.max(axis=0)

# División de los datos usando train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=73)
```

```python
out_dimension = 1
nHidden = 4
nInputDimension = X_train.shape[1]
modelo = Sequential()
modelo.add(Dense(nHidden, input_dim = nInputDimension, activation = 'relu', kernel_initializer = glorot_normal(seed=0)))
modelo.add(Dense(out_dimension, activation = 'sigmoid', kernel_initializer = glorot_normal(seed=0)))


#opt = SGD(lr=0.01)
#modelo.compile(loss = "categorical_crossentropy", optimizer = opt)

#modelo.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

#training = modelo.fit(X_train, Y_train, epochs=100)

sgd = SGD(lr=0.05, decay=1e-6, momentum=0.5, nesterov=True)
#modelo.compile(loss='sparse_categorical_crossentropy',optimizer=sgd, metrics=['accuracy'])

modelo.compile(loss='mean_squared_logarithmic_error', optimizer=sgd, metrics=['mae'])

# Last parameter (callbacks) is important for TensorBoard
train = modelo.fit(X_train, Y_train, epochs=100, callbacks=[tbCallBack])
score = modelo.evaluate(X_test, Y_test, batch_size=25)
score
```
